#**ROCK VS MINE PREDICTION**

sonar is used to send and recieve signals bounce back from metals(mines) and some rocks   
so I use this sonar data and feed this data to machine learning model(logistic regression model for binary classification) which predicts whether the object is made of metal or just rock.


**Workflow**

Sonar Data  --> Data preprocessing(standardization & vectorization)  --> train-test-split  --> Logistic Regression model{supervised learning algorithm} (which works really well for binary classification problem) --> we'll get the trained logistic regression model which will predict based on the sonar data

```python
# so we first do label encoding and then data standardization
# 3 label encoding for two labels
from sklearn.preprocessing import LabelEncoder
```

#**Handling imbalance dataset**

for making row no same for both rock and mine

```python
print(sonar_df.shape)
```

```python
# distribution of the two classes
sonar_df['target'].value_counts()
#sonar_df.target.value_counts()   is giving error remember
```

```python
# no of 0 and 1 are unequal so make them equal   so we have to make 0 sample of 96 random values
mine = sonar_df[sonar_df.target==0]
rock = sonar_df[sonar_df.target==1]
```

```python
print(mine.shape, rock.shape)
```

under-sampling

```python
mine_sample = mine.sample(n=97)
```

```python
mine_sample.shape, rock.shape
```

concatenate the two data


```python
new_dataset = pd.concat([mine_sample, rock], axis=0)
```

```python
new_dataset.head()
```

Training and test data

```python
#seperate feature and target
X = new_dataset.drop(columns=[60,'target'], axis=1)
Y = new_dataset.target
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=.1, random_state=1)   # we can simply use stratify=Y for handling imbalance dataset and save our time from handling the imbalance dataset
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
print(X)
Y
```

#**Model Training**
Logistic regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
```

```python
#training the logistic regression model with training data
model.fit(X_train, Y_train)
```

#**Model Evaluation**

```python
from sklearn.metrics import accuracy_score
```

```python
# accuracy on training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
print('Accuracy score on training data: ', training_data_accuracy*100, '%')
```

```python
# accuracy on test data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
print('Accuracy score on testing data: ', test_data_accuracy*100, '%')
```
